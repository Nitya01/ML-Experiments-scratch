# LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("airline-safety.csv")
#print(df.head())
```

```python
X = df[['fatal_accidents','incidents']].values
Y = df['fatalities'].values
m = len(df)
print(Y.shape)
```

```python
theta = np.array([[0,0]]).T
iterations = 500
alpha = 0.01
```

```python
def cost(X, Y, theta):
    m = len(X)
    J = 0
    J = (np.sum(np.power((np.dot(X, theta) - Y), 2)))/(2 * m)
    return J

cost(X, Y, theta)
```

```python
def gradient_descent_func(X, Y, theta, alpha, num_iters):
    J_his = np.zeros((num_iters,))
    T_his = np.zeros((num_iters,2))
    m = len(Y)    
    for i in range(num_iters):
        T_his[i] = theta

        J_his[i] = cost(X, Y, theta)
    return theta, J_his, T_his
```

```python
theta, J_his, T_his = gradient_descent_func(X, Y, theta, alpha, iterations)
```
