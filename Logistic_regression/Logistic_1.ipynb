```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("Product.csv")
```

```python
data
```

```python
X=data.iloc[:,[2,3]].values #slices the values of age and estimated salary values from the dataset 
```

```python
Y=data.iloc[:,4].values # purchase will be the output value
```

```python
from sklearn.preprocessing import StandardScaler #for scaling cols 2,3
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
#X_scaled
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(X[Y == 0][:, 0], X[Y == 0][:, 1], color='b', label='0')
plt.scatter(X[Y == 1][:, 0], X[Y == 1][:, 1], color='r', label='1')
plt.legend();
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y,test_size=0.2)
```

```python
np.shape(X_train)
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_predicted=model.predict(X_test)
```

```python
Y_predicted
```

```python
Y_test
```

```python
metrics.accuracy_score(Y_predicted, Y_test)
```

```python
matrix=metrics.confusion_matrix(Y_test,Y_predicted)
```

```python
matrix
```
